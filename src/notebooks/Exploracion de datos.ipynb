#### En el segundo CSV tenemos:

* Numero total de transacciones en 2021 escrituradas, en adelante  NT -- INE
* Valor total de esas transacciones, unidad de medida miles, en adelante TV -- INE
* Generare con esas columnas el valor medio de venta por provincia, en adelante VM.

```python
valor21 = pd.read_csv("../data/Num y Valor total Transaciones.csv",index_col = 0)

print( " Este sería el valor medio de Almeria >>" , int((valor21.loc['Almeria']['Valor total Transaciones']/2863) * 1000))
valor21.head()
```

```python
valor21.rename(columns = {'Valor total Transaciones': 'VT', 'Numero Transaciones' : 'NT'}, inplace = True)
```

#### Generar la columna que necesito

* VT dividido NT * 1000 nos da el valor medio por provincia

```python
valor21['VM'] = (valor21['VT'] / valor21['NT']) * 1000
valor21['VM'] = valor21['VM'].round(2)
valor21.head()
```
